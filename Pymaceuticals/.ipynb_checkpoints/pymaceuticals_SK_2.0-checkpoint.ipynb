```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem, ttest_ind

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv("data/mouse_drug_data.csv")
trial_df = pd.read_csv("data/clinicaltrial_data.csv")
```

```python
# Combine the data into a single dataset
py_df = pd.merge(trial_df, mouse_df, on="Mouse ID" , how="left")

# Display the data table for preview
py_df.head()
```

## Tumor Response to Treatment

```python
# A1: Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
py2_df = pd.DataFrame(py_df["Drug"])
py2_df["Timepoint"] = py_df["Timepoint"]
py2_df["Tumor Volume (mm3)"] = py_df["Tumor Volume (mm3)"]

# Convert to DataFrame
grouped_py = py2_df.groupby(['Drug','Timepoint'])
grouped_py = grouped_py["Tumor Volume (mm3)"].agg(['mean'])
# Preview DataFrame
grouped_py.head()
```

```python
# A2: Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
grouped_py2 = py2_df.groupby(['Drug','Timepoint'])
# Convert to DataFrame
grouped_py2 = grouped_py2["Tumor Volume (mm3)"].agg(['sem'])
# Preview DataFrame
grouped_py2.head()
```

```python
# A3: Isolate mean values for each drug 
Capomulin = pd.DataFrame(grouped_py.loc["Capomulin", "mean"])
Ceftamin = pd.DataFrame(grouped_py.loc["Ceftamin", "mean"])
Infubinol = pd.DataFrame(grouped_py.loc["Infubinol", "mean"])
Ketapril = pd.DataFrame(grouped_py.loc["Ketapril", "mean"])
Naftisol = pd.DataFrame(grouped_py.loc["Naftisol", "mean"])
Placebo = pd.DataFrame(grouped_py.loc["Placebo", "mean"])
Propriva = pd.DataFrame(grouped_py.loc["Propriva", "mean"])
Ramicane = pd.DataFrame(grouped_py.loc["Ramicane", "mean"])
Stelasyn = pd.DataFrame(grouped_py.loc["Stelasyn", "mean"])
Zoniferol = pd.DataFrame(grouped_py.loc["Zoniferol", "mean"])

# Compile chart by merging (would have been simpler with pivot function used for next chart) 
cc = pd.merge(Capomulin, Ceftamin, on="Timepoint")
cc = cc.rename(columns={"mean_x": "Capomulin", "mean_y": "Ceftamin"})
cci = pd.merge(cc, Infubinol, on="Timepoint")
cci = cci.rename(columns={"mean": "Infubinol"})
kn = pd.merge(Ketapril, Naftisol, on="Timepoint")
kn = kn.rename(columns={"mean_x": "Ketapril", "mean_y": "Naftisol"})
ccikn =  pd.merge(cci, kn, on="Timepoint")
pp = pd.merge(Placebo, Propriva, on="Timepoint")
pp = pp.rename(columns={"mean_x": "Placebo", "mean_y": "Propriva"})
ppr = pd.merge(pp, Ramicane, on="Timepoint")
ppr = ppr.rename(columns={"mean": "Ramicane"})
sz = pd.merge(Stelasyn, Zoniferol, on="Timepoint")
sz = sz.rename(columns={"mean_x": "Stelasyn", "mean_y": "Zoniferol"})
pprsz = pd.merge(ppr, sz, on="Timepoint")
cciknpprsz = pd.merge(ccikn, pprsz, on="Timepoint")
cciknpprsz
```

```python
# A4: Isolate Standard Error of Tumor Volumes  for each drug 
Capomulin_stv = pd.DataFrame(grouped_py2.loc["Capomulin", "sem"])
Ceftamin_stv = pd.DataFrame(grouped_py2.loc["Ceftamin", "sem"])
Infubinol_stv = pd.DataFrame(grouped_py2.loc["Infubinol", "sem"])
Ketapril_stv = pd.DataFrame(grouped_py2.loc["Ketapril", "sem"])
Naftisol_stv = pd.DataFrame(grouped_py2.loc["Naftisol", "sem"])
Placebo_stv = pd.DataFrame(grouped_py2.loc["Placebo", "sem"])
Propriva_stv = pd.DataFrame(grouped_py2.loc["Propriva", "sem"])
Ramicane_stv = pd.DataFrame(grouped_py2.loc["Ramicane", "sem"])
Stelasyn_stv = pd.DataFrame(grouped_py2.loc["Stelasyn", "sem"])
Zoniferol_stv = pd.DataFrame(grouped_py2.loc["Zoniferol", "sem"])
```

```python
x_tv = np.arange(0, 50, 5)
y_tvm1 = Capomulin['mean']
yerr_tvs1 = Capomulin_stv['sem']

fig = plt.figure()
plt.figure(figsize=(12,7))
plt.errorbar(x_tv, Capomulin['mean'] , yerr=Capomulin_stv['sem'], label='Capomulin')
plt.errorbar(x_tv, Ceftamin['mean'] , yerr=Ceftamin_stv['sem'], label='Ceftamin')
plt.errorbar(x_tv, Infubinol['mean'] , yerr=Infubinol_stv['sem'], label='Infubinol')
plt.errorbar(x_tv, Naftisol['mean'] , yerr=Naftisol_stv['sem'], label='Naftisol')
plt.errorbar(x_tv, Placebo['mean'] , yerr=Placebo_stv['sem'], label='Placebo')
plt.errorbar(x_tv, Propriva['mean'] , yerr=Propriva_stv['sem'], label='Propriva')
plt.errorbar(x_tv, Ramicane['mean'] , yerr=Ramicane_stv['sem'], label='Ramicane')
plt.errorbar(x_tv, Stelasyn['mean'] , yerr=Stelasyn_stv['sem'], label='Stelasyn')
plt.errorbar(x_tv, Zoniferol['mean'] , yerr=Zoniferol_stv['sem'], label='Zoniferol')
plt.title("Tumor Response to Treatment")
plt.legend()
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Timepoint")

```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
py3_df = pd.DataFrame(py_df["Drug"])
py3_df["Timepoint"] = py_df["Timepoint"]
py3_df["Metastatic Sites"] = py_df["Metastatic Sites"]

# Convert to DataFrame
grouped_py3 = py3_df.groupby(['Drug','Timepoint'])
grouped_py3 = grouped_py3["Metastatic Sites"].agg(['mean'])
# Preview DataFrame
grouped_py3.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
grouped_py4 = py3_df.groupby(['Drug','Timepoint'])
# Convert to DataFrame
grouped_py4 = grouped_py4["Metastatic Sites"].agg(['sem'])
# Preview DataFrame
grouped_py4.head()
```

```python
pyt_df = grouped_py3.reset_index()
py3_pv = pyt_df.pivot(index="Timepoint", columns="Drug")["mean"]
py3_pv
```

```python
pyt4_df = grouped_py4.reset_index()
py4_pv = pyt4_df.pivot(index="Timepoint", columns="Drug")["sem"]
py4_pv
```

```python
# Generate the Plot (with Error Bars)
x_tv = np.arange(0, 50, 5)

fig = plt.figure()
plt.figure(figsize=(10,7))
plt.errorbar(x_tv, py3_pv['Capomulin'] , yerr=py4_pv['Capomulin']/2, label='Capomulin' )
plt.errorbar(x_tv, py3_pv['Ceftamin'] , yerr=py4_pv['Ceftamin']/2, label='Ceftamin')
plt.errorbar(x_tv, py3_pv['Infubinol'] , yerr=py4_pv['Infubinol']/2, label='Infubinol')
plt.errorbar(x_tv, py3_pv['Ketapril'] , yerr=py4_pv['Ketapril']/2, label='Ketapril')
plt.errorbar(x_tv, py3_pv['Naftisol'] , yerr=py4_pv['Naftisol']/2, label='Naftisol')
plt.errorbar(x_tv, py3_pv['Placebo'] , yerr=py4_pv['Placebo']/2, label='Placebo')
plt.errorbar(x_tv, py3_pv['Propriva'] , yerr=py4_pv['Propriva']/2, label='Propriva')
plt.errorbar(x_tv, py3_pv['Ramicane'] , yerr=py4_pv['Ramicane']/2, label='Ramicane')
plt.errorbar(x_tv, py3_pv['Stelasyn'] , yerr=py4_pv['Stelasyn']/2, label='Stelasyn')
plt.errorbar(x_tv, py3_pv['Zoniferol'] , yerr=py4_pv['Zoniferol']/2, label='Zoniferol')
plt.title("Metastatic Spread During Treatment")
plt.legend()
plt.ylabel("Met. Sites")
plt.xlabel("Timepoint")
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
# Convert to DataFrame
# Preview DataFrame
py5_df = pd.DataFrame(py_df["Drug"])
py5_df["Timepoint"] = py_df["Timepoint"]
py5_df["Mouse ID"] = py_df["Mouse ID"]

# Convert to DataFrame
grouped_py5 = py5_df.groupby(['Drug','Timepoint'])
grouped_py5 = grouped_py5["Mouse ID"].agg(['count'])
# Preview DataFrame
grouped_py5.head()
```

```python
pyt5_df = grouped_py5.reset_index()
py5_pv = pyt5_df.pivot(index="Timepoint", columns="Drug")["count"]
py5_pv
```

```python
# Generate the Plot (Accounting for percentages)
fig = plt.figure()
plt.figure(figsize=(12,7))
plt.ylabel("Survival Rate(%)")
plt.xlabel("Time(Days)")
plt.title("Survival Rates During Treatment")

# Create a handle for each plot
Capomulin_sr, = plt.plot(py5_pv.iloc[:,0]/py5_pv.iloc[0,0]*100, marker="+",color="blue", linewidth=1, label="Capomulin")
Ceftamin_sr, = plt.plot(py5_pv.iloc[:,1]/py5_pv.iloc[0,1]*100, marker="s",color="green", linewidth=1, label="Ceftamin")
Infubinol_sr, = plt.plot(py5_pv.iloc[:,2]/py5_pv.iloc[0,2]*100, marker="+",color="red", linewidth=1, label="Infubinol")
Ketapril_sr, = plt.plot(py5_pv.iloc[:,3]/py5_pv.iloc[0,3]*100, marker="s",color="yellow", linewidth=1, label="Ketapril")
Naftisol_sr, = plt.plot(py5_pv.iloc[:,4]/py5_pv.iloc[0,4]*100, marker="+",color="pink", linewidth=1, label="Naftisol")
Placebo_sr, = plt.plot(py5_pv.iloc[:,5]/py5_pv.iloc[0,5]*100, marker="s",color="blue", linewidth=1, label="Placebo")
Propriva_sr, = plt.plot(py5_pv.iloc[:,6]/py5_pv.iloc[0,6]*100, marker="+",color="green", linewidth=1, label="Propriva")
Ramicane_sr, = plt.plot(py5_pv.iloc[:,7]/py5_pv.iloc[0,7]*100, marker="s",color="red", linewidth=1, label="Ramicane")
Stelasyn_sr, = plt.plot(py5_pv.iloc[:,8]/py5_pv.iloc[0,8]*100, marker="+",color="yellow", linewidth=1, label="Stelasyn")
Zoniferol_sr, = plt.plot(py5_pv.iloc[:,9]/py5_pv.iloc[0,9]*100, marker="s",color="orange", linewidth=1, label="Zoniferol")

# Set our legend to where the chart thinks is best
plt.legend(handles=[Capomulin_sr, Ceftamin_sr, Infubinol_sr, Ketapril_sr, Naftisol_sr, Placebo_sr, 
                    Propriva_sr, Ramicane_sr, Stelasyn_sr, Zoniferol_sr], loc="best")
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
drug_change = cciknpprsz
drug_change_list =[]
for x in range(0, 10):
    change =(drug_change.iloc[9,x] - drug_change.iloc[0,x]) / drug_change.iloc[0,x]*100
    drug_change_list.append(change)

# Create dataframe for percent changes for each drug    
drug_change_percent = pd.DataFrame()    
drug_change_percent["Drug"] = drug_change.columns 
drug_change_percent["Percent Change"] = drug_change_list
drug_change_percent
```

```python
# Store all Relevant Percent Changes into a Tuple
percent_tuple = tuple(drug_change_percent.itertuples(index=False, name=None))
percent_tuple

# Splice the data between passing and failing drugs
sorted_drugs = sorted(percent_tuple, key = lambda x: float(x[1]), reverse = True) 
passing_drugs = sorted_drugs[8:10]
failing_drugs = sorted_drugs[0:7]
print(f"Passing drugs: {passing_drugs}")
print(f"Failing drugs: {failing_drugs}")

# Create Chart  
fig = plt.figure()
plt.figure(figsize=(10,7))
data = drug_change_percent["Percent Change"]
x_axis = drug_change_percent["Drug"]
plt.bar(x_axis, data, color="b", align="center")
```
